# Data Wrangling Notebook

This notebook is for combining and cleaning datasets for crime statistics from https://www.ojjdp.gov/ojstatbb/crime/ucr_trend.asp?table_in=2 into a data frame for crime by type with related demographic data. Released onNovember 16, 2020

Some datasets are for arrest rates per 100,000 population units. Others are counts:
* _rates = per 100,000 population units os specific crime types
* _ucr = total violent crimes for demographic attribures


```python
import numpy as np
import pandas as pd
import glob
```

```python
#Load all files into one dataframe.
files = !ls *1.csv # IPython magic
d = pd.concat([pd.read_csv(f, index_col=None, skiprows=1, header=1, nrows=40) for f in files], axis=0, keys=files)
d
```

```python
#Cleanup 'Crime' column.
d.Crime.fillna(d.crime, inplace=True)
del d['crime']
d
```

```python
d.replace(',','', regex=True, inplace=True)
```

```python
d.to_csv(r'crime_rates_by_type.csv', index = False)
```
